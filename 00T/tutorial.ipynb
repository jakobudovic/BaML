# Exercise T0.1 Oktoberfest Beer Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Oktoberfest.csv")
df.head()
```

```python
# b) Take a look at the dataframeâ€™s column names.
list(df.columns)
```

```python
# c) What was the price of a beer in 1995?
float(df.loc[df["Year"] == 1995].Beer_Price)
```

```python
# d) Which year is the earliest recorded entry of the dataset?
df["Year"].sort_values(ascending=True)[0]
```

```python
# e) Determine the value range of "Visitors_Total".
df.Visitors_Total.min(), df.Visitors_Total.max()
```

```python
# f) Plot the beer consumption over the years.
fig, ax = plt.subplots()

ax.plot(df.Year, df.Beer_Consumption)

ax.set(xlabel='years', ylabel='consumption (L)',
       title='Beer consumption over the years')
ax.grid()

plt.show()
```

```python
# alternative
df.plot(x="Year", y=["Beer_Consumption", "Beer_Price"])
```

```python
# g) Create a scatterplot that shows the number of visitors per year. Calculate the correlation between "Visitors_Total" and "Beer_Consumption".
fig, ax = plt.subplots()

# get data
years = df.Year
visitors = df.Visitors_Total

ax.scatter(years, visitors)

ax.set(xlabel='years', ylabel='visitors',
       title='Visitors over the years')
ax.grid()
plt.show()
```

```python
df.Visitors_Total.corr(df.Beer_Consumption)
```

```python
# h) Compute the mean and the variance of beer price during the period from 2000 to 2007.
beer_price = df[(df["Year"] >= 2000) & (df["Year"] <= 2007)].Beer_Price
print(f"Mean price: {beer_price.mean()}")
print(f"Variance of the price: {beer_price.var():0.3}")
```

```python
# alternative
df[df["Year"].between(2000, 2007)]
```

```python
# i) Add a new column to the dataframe, describing the difference of beer prices between two consecutive years. Plot this difference against "Year".
df["Beer_Price_Difference"] = df.diff().Beer_Price
df["Beer_Price_Difference"].replace(np.nan, 0, inplace=True)
df.head()
```

```python
df.plot(x="Year", y="Beer_Price_Difference")
```
